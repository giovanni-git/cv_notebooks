## Filtri di Gabor

Il filtro di Gabor (in forma semplificata) è definito dalla seguente funzione che restituisce due parti (reale e immaginaria)


$$RE_g(x,y,\omega,\theta) = \frac{\omega^2}{4\pi^3}\left[\exp\left\{-(\omega^2/8\pi^2) (4x'^2 + y'^2)\right\}\right]\exp\left(\pi^2/2\right)(\omega x \cos \theta)$$

$$IM_g(x,y,\omega,\theta) = \frac{\omega^2}{4\pi^3}\left[\exp\left\{-(\omega^2/8\pi^2) (4x'^2 + y'^2)\right\}\right]\exp\left(\pi^2/2\right)(\omega x \sin \theta)$$

dove $x' = x \cos \theta + y \sin \theta$ e $y' = y \cos \theta. - x \sin \theta$. L'implementazione è data dal codice sottostante: 

```python
zebra =  rgb2gray(imageio.imread('zebra.jpg'))
```

```python
myimshow(zebra)
```

```python
def genGabor(sz, omega, theta, func=np.cos):
    radius = (int(sz[0]/2.0), int(sz[1]/2.0))
    [x, y] = np.meshgrid(range(-radius[0], radius[0]), range(-radius[1], radius[1]))

    x1 = x * np.cos(theta) + y * np.sin(theta)
    y1 = -x * np.sin(theta) + y * np.cos(theta)
    
    gauss = omega**2 / (4*np.pi**3) * np.exp(- omega**2 / (8*np.pi**2) * ( 4 * x1**2 + y1**2))
    sinusoid = func(omega * x1) * np.exp(np.pi**2 / 2)
    gabor = gauss * sinusoid
    return gabor
```

`sz` rappresenta la dimensione del filtro, `omega` la frequenza e `theta` l'orientamento.

```python

def hough_lines(image):
    img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    img = skimage.img_as_ubyte(img_gray)
    img_gray_blur = cv2.GaussianBlur(img, (11,11), 5)
    img_edges=cv2.Canny(img_gray_blur,10,70,None,3)
    lines=cv2.HoughLinesP(img_edges,1,np.pi/180,125,minLineLength=50,maxLineGap=250)

    print("numero di linee",len(lines))
    for line in lines:
        
        x1,y1,x2,y2=line[0]
        cv2.line(image,(x1,y1),(x2,y2),(0,0,255),3)
    return image

def draw_contours(image):
    img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    img = skimage.img_as_ubyte(img_gray)
    img_gray_blur = cv2.GaussianBlur(img, (11,11), 3)
    thresh = cv2.Canny(img_gray_blur, 10, 70, None, 3)
    contours, hierarchy = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    print("numero di contorni",len(contours))
    for i in range(len(contours)):
        cv2.drawContours(image, contours, i, (0, 255, 0), 3)

    return image;

plt.figure(2,figsize=(10,20))
strisce=cv2.imread('gabor1.jpg')
img_lines=hough_lines(strisce)
img_contourn=draw_contours(strisce)
plt.subplot(121);plt.imshow(img_lines)

plt.subplot(122);plt.imshow(img_contourn)
plt.show()
```

Identificare il filtro e riprodurre lo sketch in figura

![](model1.jpg)
![](model1_post.jpg)

```python
def sketch(image):
    img_gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    img_gray_blur=cv2.GaussianBlur(img_gray,(5,5),7)
    canny_edges=cv2.Canny(img_gray_blur,40,70,None,3)
    ret,mask=cv2.threshold(canny_edges,70,255,cv2.THRESH_BINARY_INV)
    return mask

model=cv2.imread('model1.jpg')

plt.figure(2,figsize=(10,30))
plt.subplot(121);plt.imshow(model)
sketched_image=sketch(model)
plt.subplot(122);plt.imshow(sketched_image)
plt.show()
```
